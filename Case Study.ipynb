```python
class Vehicle:
    def __init__(self, vehicle_type):
        self.vehicle_type = vehicle_type

    def display(self):
        print(f"  Vehicle Type : {self.vehicle_type.capitalize()}")


class Automobile(Vehicle):
    def __init__(self, vehicle_type, year, make, model, doors, roof):
        super().__init__(vehicle_type)
        self.year  = year
        self.make  = make
        self.model = model
        self.doors = doors
        self.roof  = roof

    def display(self):
        print("\n" + "=" * 35)
        print("       VEHICLE INFORMATION")
        print("=" * 35)
        super().display()
        print(f"  Year         : {self.year}")
        print(f"  Make         : {self.make}")
        print(f"  Model        : {self.model}")
        print(f"  Doors        : {self.doors}")
        print(f"  Roof         : {self.roof}")
        print("=" * 35 + "\n")
```

```python
vehicle_type = "car"
print(f'Vehicle type is set to: "{vehicle_type}"\n')

year  = input("Enter the year   : ").strip()
make  = input("Enter the make   : ").strip()
model = input("Enter the model  : ").strip()

while True:
    doors = input("Enter number of doors (2 or 4): ").strip()
    if doors in ["2", "4"]:
        break
    print("  Please enter 2 or 4.")

while True:
    roof = input('Enter type of roof - "solid" or "sun roof": ').strip()
    if roof in ["solid", "sun roof"]:
        break
    print('  Please enter "solid" or "sun roof".')

my_car = Automobile(vehicle_type, year, make, model, doors, roof)
my_car.display()
```
